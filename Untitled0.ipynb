```python
# import statements
from google.colab import drive
from matplotlib import pyplot as plt
import numpy as np
import os
```

```python
# set up access to google drive
drive.mount('/content/gdrive')
gdrive_directory = '/content/gdrive/MyDrive/Adaptive Scanning'

os.chdir(gdrive_directory)
if os.path.exists(gdrive_directory):
  print('Path exists')
else:
  print('Path does not exist')
```

```python
# file name option lists
img_name_options=["aphid","debris","forceps"]
mode_options=["random","random-shift-random","grid-shift-path","grid-shift-structured","grid-shift-random"]
solver_options=["x-order","nn-order","lk-order-cycle"]
optimised_segments_options=[False, True]
```

```python
# nested for loops for reading from file structure
for img_name in img_name_options:
  for optimised_segments in optimised_segments_options:
    for solver in solver_options:
      for mode in mode_options:
        for counter in range(5):
          internal_directory = f"{img_name}/Test Data/{optimised_segments}/{solver}/{mode}/{counter}/"
          final_directory = os.path.join(gdrive_directory, internal_directory)
          if os.path.exists(final_directory):
            os.chdir(final_directory)
            transit_history = np.load(os.path.join(final_directory, 'scan-transit-history.npy'))
            print(transit_history)
          else:
            print(f'Path {final_directory} does not exist')
                        
```
